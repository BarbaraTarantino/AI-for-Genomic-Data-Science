# Pandas, Matplotlib, and Seaborn – Task Notebook

In this notebook, we will explore the fundamental concepts for data manipulation and visualization in Python, using three main libraries:

- **Pandas**: for managing and analyzing tabular data;
- **Matplotlib**: for creating basic and multiple plots;
- **Seaborn**: for elegant and effective statistical visualizations.

Each section is organized into **numbered Tasks**, each containing: an introductory explanation, example code, and interpretation of the results.

```python
# Environment Setup

#!pip install pandas matplotlib seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Environment ready — Pandas, Matplotlib, and Seaborn installed.")
```

## Part 1 – Pandas

### Task 1 – Creation of a DataFrame

With Pandas, we can create a **DataFrame**, a tabular structure that organizes data into rows and columns.

```python
data = {
    "Sample": ["S01", "S02", "S03", "S04", "S05", "S06"],
    "Type": ["ALL", "ALL", "AML", "AML", "ALL", "AML"],
    "GATA1": [5.2, 5.5, 6.1, 5.9, 5.0, 6.3],
    "MYC": [6.0, 5.8, 6.5, 6.7, 5.9, 6.2],
    "TP53": [4.9, 5.1, 5.3, 5.4, 5.0, 5.2]
}

df = pd.DataFrame(data)
print(df)
```

We created a DataFrame with 6 samples, the type of leukemia, and the expression values of three genes.

### Task 2 – Data Inspection

To better understand the dataset, we can inspect its first few rows, data types, and descriptive statistics.

```python
print(df.head())        # First rows
```

```python
print(df.info())        # Data type info
```

```python
print(df.describe())    # Descriptive statistics
```

We obtain a summary of the dataset: data types, memory usage, and statistics on the numerical variables.

### Task 3 – Selection of Columns and Rows

It is often useful to select specific columns, rows, or subsets of data for targeted analysis.

```python
print(df["MYC"])        # Single column
```

```python
print(df.loc[0])        # First row by label
```

```python
print(df.iloc[0:3, 2:4]) # Sub-table using slicing
```

We can easily select columns, rows, or subsets of data.

### Task 4 – Data Filtering

With Pandas, we can filter data based on logical conditions, for example by selecting only a subset of samples.

```python
aml = df[df["Type"] == "AML"]
print(aml)
```

We filtered only the samples belonging to the AML subtype.

### Task 5 – Data Operations

Pandas allows for the execution of statistical operations and data groupings in a simple and efficient manner.

```python
print("Mean MYC:", df["MYC"].mean())
```

```python
print("\nMean by Type:")
print(df.groupby("Type")["MYC"].mean())
```

Let's calculate the overall average of MYC and the average for each subtype of leukemia.

## Part 2 – Matplotlib

### Task 6 – Line Chart

Matplotlib allows for the graphical representation of the trend of a variable with respect to an index or category.

```python
# A single graph: Axes created automatically
plt.plot(df["Sample"], df["GATA1"], marker="o")
plt.title("GATA1 Expression")
plt.xlabel("Sample")
plt.ylabel("log2 Value")
plt.show()
```

The graph shows the trend of GATA1 values in the different samples.

### Task 7 – Histogram

A histogram shows how the values of a variable are distributed within the dataset.

```python
plt.hist(df["MYC"], bins=5, color="skyblue", edgecolor="black")
plt.title("MYC Distribution")
plt.xlabel("log2 Expression")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows how the expression values of the MYC gene are distributed.

### Task 8 – Scatter Plot

A scatter plot allows for the visualization of the relationship between two numerical variables.

```python
plt.scatter(df["GATA1"], df["MYC"], color="purple")
plt.title("Scatter Plot: GATA1 vs MYC")
plt.xlabel("GATA1")
plt.ylabel("MYC")
plt.show()
```

The points highlight the relationship between the values of GATA1 and MYC.

### Task 9 – Boxplot

A boxplot highlights the distribution of a variable by displaying the median, quartiles, and potential outliers.

```python
plt.boxplot(df["TP53"])
plt.title("TP53 Boxplot")
plt.show()
```

The boxplot displays the median, quartiles, and potential outliers for the TP53 values.

### Task 10 – Multiple Graphs with Matplotlib

In this exercise, we will see how to create multiple graphs within the same figure using the function `plt.subplots()`.

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(df["GATA1"], bins=5, color="orange")
axes[0].set_title("GATA1 Distribution")

axes[1].hist(df["TP53"], bins=5, color="green")
axes[1].set_title("TP53 Distribution")

plt.tight_layout()
plt.show()
```

Let's take a quick look at the distributions of GATA1 and TP53.

## Part 3 – Seaborn

### Task 11 – Scatter Plot with Categories

In this exercise, we explore how to create scatter plots using Seaborn, utilizing the `hue` parameter to color the points based on a categorical variable.

```python
sns.scatterplot(x="GATA1", y="MYC", hue="Type", data=df)
plt.title("GATA1 vs MYC by Type")
plt.show()
```

The points are colored according to the subtype (ALL or AML).

### Task 12 – Boxplot by Categories

In this example, we analyze the distribution of a gene (MYC) across different subtypes of leukemia (ALL vs AML).

```python
sns.boxplot(x="Type", y="MYC", data=df)
plt.title("MYC Expression by Type")
plt.show()
```

The boxplot compares the distribution of MYC values between the two subtypes.

### Task 13 – Correlation Heatmap

Here we visualize the correlation matrix between genes using a heatmap, which immediately displays the strength of the relationships between numerical variables.

```python
corr = df[["GATA1", "MYC", "TP53"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix among Genes")
plt.show()
```

The heatmap highlights the correlation coefficients between the genes GATA1, MYC, and TP53.

### Task 14 – Multiple Graphs with Seaborn

In this exercise, we utilize Figure-level functions, which automatically create and manage figures with multiple subplots.

- With **`FacetGrid`**, we visualize the *distribution of MYC* separated by subtype (ALL vs AML).  
- With **`pairplot`**, we explore the *relationships among multiple genes* (GATA1, MYC, TP53) in a single figure with scatter plots and histograms.

The Figure-level functions manage the entire figure, not a single axis:  
`FacetGrid` / `PairGrid` → multiple automatic axes → enhanced statistical comparison.

```python
# Multiple Distributions with FacetGrid
g = sns.FacetGrid(df, col="Type")
g.map_dataframe(sns.histplot, x="MYC", bins=5, color="skyblue")
plt.show()
```

```python
# Multiple scatter plots with pairplot
sns.pairplot(df[["GATA1", "MYC", "TP53", "Type"]], hue="Type")
plt.show()
```

With `FacetGrid`, we observe the distribution of MYC separated by subtype. With `pairplot`, we simultaneously examine the relationships among multiple genes.

### Final Summary: Axis-level vs Figure-level

| Function Type   | Control      | Examples                          | Characteristic                                      |
|------------------|--------------|-----------------------------------|-----------------------------------------------------|
| **Axis-level**    | Single axis (`ax`) | `sns.boxplot()`, `sns.heatmap()` | Integrates into already created figures, uses `plt.subplots()` |
| **Figure-level**  | Complete figure | `sns.relplot()`, `sns.catplot()`, `sns.FacetGrid()`, `sns.pairplot()` | Automatically creates and manages multiple figures   |
